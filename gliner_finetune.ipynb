```python
import json
import pandas as pd


df = pd.read_csv("../dataset/dataset.csv")
# select a sample fo rows
df = df.sample(200)
df.head()
```

```python
import json
import ast

ft_dataset = []
for r in df.to_dict(orient='records'):
    patterns = []
    try:
        for pm in ast.literal_eval(r["privacy_mask"]):
            patterns.append({"label": pm['label'], "pattern": pm['value']})
    except:
        print(json.dumps(r, indent=4))
        raise Exception()
    ft_dataset.append({
        "text": r['source_text'], 
        "patterns": patterns,
        "lang": r["language"]
        })
```

```python
import spacy
from sklearn.model_selection import train_test_split
from pathlib import Path
import copy
from datetime import datetime


def save_data(data, file_path, overwrite):
    """Save data to a file, handling overwriting based on user preference."""
    path = Path(file_path)
    assets_dir = path.parent
    if not assets_dir.exists():
        assets_dir.mkdir(parents=True, exist_ok=True)
    
    if not overwrite and path.exists():
        timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
        file_path = f"{path.stem}_{timestamp}{path.suffix}"
    
    with open(file_path, 'w') as f:
        json.dump(data, f)
    
    print(f"Data saved to {file_path}")

# def create_patterns2(example):
#     patterns = []
#     for category, items in example.items():
#         if isinstance(items, list) and category != 'text':
#             for item in items:
#                 patterns.append({"label": category.upper(), "pattern": item})
#     return patterns
# #[{'label': 'GENERIC_PLANT_FOOD', 'pattern': 'grasses'}, {'label': 'PLANT_FOOD', 'pattern': 'leaves'}, {'label': 'PLANT_FOOD', 'pattern': 'bark'}, {'label': 'LOCATION_NEST', 'pattern': 'large trees'}, {'label': 'ITEM_NEST', 'pattern': 'mud'}, {'label': 'ITEM_NEST', 'pattern': 'sand'}]


TOKENIZER_SUPPORTED_LANGUAGES = ["en", "fr", "it", "es", "nl"] 

""" manually override functions """
def extract_annotations(text, patterns, lang_model):
    # asse rt tokenizer_lang in TOKENIZER_SUPPORTED_LANGUAGES, f"Tokenizer language {tokenizer_lang} not supported. Supported languages: {TOKENIZER_SUPPORTED_LNAGUAGES}"
    assert lang_model is not None, "Language model for the provided language is not found. Please provide a valid language."

    nlp = copy.deepcopy(lang_model)
    ruler = nlp.add_pipe("entity_ruler")
    ruler.add_patterns(patterns)

    # Process the text
    doc = nlp(text)

    # Preparing the output format
    # TODO: use char position instead of text
    tokenized_text = [token.text for token in doc]
    ner = []
    for ent in doc.ents:
        start = ent.start
        end = ent.end-1  # Adjusting end index to be inclusive, not +1
        ner.append([start, end, ent.label_])

    return {"tokenized_text": tokenized_text, "ner": ner}

# {'tokenized_text': ['The', 'African', 'Elephant', 'primarily', 'consumes', 'grasses', ',', 'leaves', ',', 'and', 'bark', '.', 'It', 'gathers', 'its', 'food', 'by', 'pulling', 'it', 'down', 'with', 'its', 'trunk', 'while', 'roaming', 'through', 'the', 'African', 'plains', '.', 'It', 'rests', 'under', 'large', 'trees', 'and', 'uses', 'mud', 'and', 'sand', 'to', 'create', 'a', 'cool', 'and', 'comfortable', 'sleeping', 'environment', '.'], 'ner': [[5, 5, 'GENERIC_PLANT_FOOD'], [7, 7, 'PLANT_FOOD'], [10, 10, 'PLANT_FOOD'], [33, 34, 'LOCATION_NEST'], [37, 37, 'ITEM_NEST'], [39, 39, 'ITEM_NEST']]}


def train_eval_test_split(data_patterns, eval_split=0.2, test_split=0.0, train_file="train.json", eval_file="eval.json", test_file="test.json", overwrite=True, random_state=42):
    """Process data and split into training, validation, and testing datasets."""

    models_lang = dict()
    for lang in TOKENIZER_SUPPORTED_LANGUAGES:
        models_lang[lang] = spacy.blank(lang)

    training_data = [extract_annotations(d["text"], d["patterns"], models_lang.get(d["lang"], None)) for d in data_patterns]

    # Handle the data splitting
    if test_split > 0:
        train_val, test = train_test_split(training_data, test_size=test_split, random_state=random_state)
        save_data(test, Path('assets', test_file), overwrite)
    else:
        train_val = training_data

    eval_size = eval_split / (1 - test_split)  # Adjust eval size based on the remaining data
    train, val = train_test_split(train_val, test_size=eval_size, random_state=random_state)

    # Save the data
    save_data(train, Path('assets', train_file), overwrite)
    save_data(val, Path('assets', eval_file), overwrite)

    return training_data
```
